```python
# %load rapid_imports.py
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
tips = pd.read_csv('https://raw.githubusercontent.com/wesm/pydata-book/2nd-edition/examples/tips.csv')
tips.head()
```

```python
def add_random_stuff(df, col, val):
    df_suffix = df.sample(10)
    df_suffix[col] = val
    return df.append(df_suffix, ignore_index=True)

def sample_per_category(df, n):
    print(f'{len(df) = }')
    return df.sample(n)

(
    tips
    [lambda df: df['tip'] > 1.5]
    .assign(tip_deviation=lambda df:
            df['tip'] - df['tip'].median())
    .pipe(add_random_stuff, 'time', 'Breakfast')
    .groupby('time', group_keys=False)
    .apply(sample_per_category, 3)
    .reset_index(drop=True)
    .assign(mean_tip_per_time=lambda df:
            df.groupby('time')
            ['tip']
            .transform(lambda s: s.mean())
            .round(2)
            )
)
```

```python
tips['tip_pct'] = tips.tip/tips.total_bill
```

```python
grouped = tips.groupby(['day', 'smoker'])
```

```python
grouped['total_bill', 'tip_pct'].agg([('avg', 'mean'), ('std', np.std)])['tip_pct']
```
