## Seaborn

```python
import seaborn as sns
```

```python
tips = pd.read_csv('https://raw.githubusercontent.com/wesm/pydata-book/2nd-edition/examples/tips.csv')
tips.head()
```

```python
party_counts = pd.crosstab(tips['day'], tips['size'])
party_counts = party_counts.loc[:, 2:5]
party_counts.head()
```

```python
with sns.axes_style('whitegrid'):
    fig, ax = plt.subplots(figsize=[6,4])
    party_counts.plot.bar(rot=30, ax=ax)
```

```python
fig, ax = plt.subplots(figsize=[6,4])
sns.barplot(x='size', y='day', hue='time', data=tips, orient='h')
```

### Multiple plots

```python
with sns.axes_style('whitegrid'):
    sns.catplot(x='day', y='tip',
                col='smoker', row='time', 
                data=tips, 
                kind='box', 
                order=['Thur', 'Fri', 'Sat', 'Sun'],
                height=4, aspect=1
               )
```

```python
sns.displot(x='tip', 
            data=tips, 
            kde=True, 
            col='smoker', 
            hue='time', hue_order=['Lunch', 'Dinner'], 
            alpha = 0.8)
```
